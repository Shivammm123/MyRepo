```python
import datetime
import matplotlib.pyplot as plt
from timestamp import convert_timestamp
from timestamp import main
from timestamp import plot_timestamps
```

```python
unix_timestamps = [1725279018, 1725279046, 1725279074, 1725279122, 1725279521, 1725285560, 1725285580,1725285611, 1725285634, 1725287847, 1725291850, 1725291868, 1725293059, 1725293738,1725293851, 1725296285, 1725296523, 1725296534, 1725296580, 1725296614, 1725296643,1725297319, 1725297555, 1725297628, 1725297736, 1725298908, 1725300365, 1725303944, 1725321256, 1725322717, 1725326995, 1725336541, 1725343605, 1725344318, 1725345532,1725353602, 1725362820]
human_readable_dates = [convert_timestamp(ts) for ts in unix_timestamps]
for timestamp, human_readable_date in zip(unix_timestamps, human_readable_dates):
    print(f"Unix Timestamp: {timestamp} -> Human-Readable Time: {human_readable_date}")
```

```python
elapsed_hours = [(ts - unix_timestamps[0]) / 3600 for ts in unix_timestamps]
plt.figure(figsize=(10, 6))
plt.plot(elapsed_hours, human_readable_dates, marker='o', linestyle='-', color='b')
plt.xlabel('Hours from First Timestamp')
plt.ylabel('Converted Time (YYYY-MM-DD HH:MM:SS)')
plt.title('Converted Times as a Function of Hours from the First Timestamp')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
hours = [int(elapsed_hour) for elapsed_hour in elapsed_hours]
plt.figure(figsize=(10, 6))
plt.hist(hours, bins=range(min(hours), max(hours) + 2), edgecolor='black', alpha=0.7)
plt.xlabel('Elapsed Hours from First Timestamp')
plt.ylabel('Number of Timestamps')
plt.title('Histogram of Timestamps by Elapsed Hours')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.hist(hours, bins=range(0, 25), edgecolor='black', alpha=0.7, label='Histogram')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Timestamps')
plt.title('Histogram of Timestamps by Hour')
plt.grid(True)
plt.xticks(range(0, 24))
plt.scatter(hours, [1]*len(hours), color='red', label='Scatter Plot')
plt.legend()
plt.show()
```
