```python
First_Repeat = np.max(np.abs(np.asarray(data1["angular_velocity"])[:, 50:80]), axis=1)
Second_Repeat= np.max(np.abs(np.asarray(data2["angular_velocity"])[:, 50:80]), axis=1)
Third_Repeat = np.max(np.abs(np.asarray(data3["angular_velocity"])[:, 50:80]), axis=1)
```

```python
Mean_J73 = np.mean(AngV_J73)
print(Mean_J73)
std_dev = np.std(AngV_J73, ddof=1)  
std_error = std_dev / np.sqrt(len(AngV_J73))
print(f"Standard Deviation of J73: {std_dev:.4f}")
print(f"Standard Error of J73: {std_error:.4f}")
```

```python
J73_Max_AngVel = np.concatenate((First_Repeat, Second_Repeat, Third_Repeat))
mean = np.mean(J73_Max_AngVel)
std_dev = np.std(J73_Max_AngVel, ddof=1)
std_err = std_dev / np.sqrt(len(J73_Max_AngVel))
```

```python
shapiro_stat, shapiro_p = stats.shapiro(J73_Max_AngVel)
dagostino_stat, dagostino_p = stats.normaltest(J73_Max_AngVel)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 0].hist(J73_Max_AngVel, bins=30, edgecolor='black')
axs[0, 0].set_title('Frequency plot of J73_Max_AngVel')

axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Frequency')

stats.probplot(J73_Max_AngVel, dist="norm", plot=axs[0, 1])
axs[0, 1].set_title('Q-Q Normality Plot of J73_Max_AngVel')

axs[1, 0].text(0.5, 0.8, f'J73 Max_AngVel Mean: {mean:.2f}', fontsize=12, ha='center')
axs[1, 0].text(0.5, 0.6, f'J73 Max_AngVel Standard Deviation: {std_dev:.2f}', fontsize=12, ha='center')
axs[1, 0].text(0.5, 0.4, f'J73 Max_AngVel Standard Error: {std_err:.2f}', fontsize=12, ha='center')
axs[1, 0].set_title('Statistics from J73_Max_AngVel')
axs[1, 0].axis('off')

axs[1, 1].text(0.5, 0.8, f'J73 Max_AngVel Shapiro-Wilk Test Statistic: {shapiro_stat:.3f}', fontsize=12, ha='center')
axs[1, 1].text(0.5, 0.6, f'p-value: {shapiro_p:.3e}', fontsize=12, ha='center')
axs[1, 1].text(0.5, 0.4, f"J73 Max_AngVel D'Agostino's K-squared Test Statistic: {dagostino_stat:.3f}", fontsize=12, ha='center')
axs[1, 1].text(0.5, 0.2, f'p-value: {dagostino_p:.3e}', fontsize=12, ha='center')
axs[1, 1].set_title('Normality Tests for J73_Max_AngVel')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()






```
